```python
import pandas as pd
```

```python
df1=pd.DataFrame({"clg":["nec","nit","nips",],"temperature":[44,40,35],"humb":[23,25,35]})
df1
```

```python
df2=pd.DataFrame({"clg":["giet","gier","gips",],"temperature":[45,45,25],"humb":[28,15,95]})
df2
```

```python
df=pd.concat([df1,df2])
df
```

```python
df=pd.concat([df1,df2],keys=["n1","n2"])
df
```

## Group by
